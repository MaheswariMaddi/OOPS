```python
table = []
#creating a class
class Student:
    #doc string
    """
        Student Class data type
    """
    #Optional or deafault Attribute
    schoolName = 'Global Minds'
    #creating constructor
    def __init__(self, name, age, gender, rollNumber):
        self.name = name #Instance or object attribute
        self.age = age
        self.gender = gender

        self.setRollNumber(rollNumber)
        self.addStudent(rollNumber, name)

    #object(or)instance method
    def addStudent(self,id, name):
        student = {'id': id, 'name': name}
        table.append(student)


    def setRollNumber(self, rollNumber):
        for student in table:
            if(student['id'] == rollNumber):
                raise f"RollNumber - {rollNumber} already exist."
        else:
            self.rollNumber = rollNumber


    def setScores(self, m1, m2):
        for student in table:
            if(student['id'] == self.rollNumber):
                student['M1'] = m1
                student['M2'] = m2


    def getStudentTotalScore(self):
        for student in table:
            if(student['id'] == self.rollNumber):
                return  student['M1'] + student['M2']

    #this method will execute when object is called.
    def __str__(self):
        return f'''
                Name - {self.name}
                Age - {self.age} 
                Gender - {self.gender}
                '''
    
ravi = Student("Ravi", 14, 'M', 1)
sneha = Student("Sneha", 13, 'F', 2)
Lenin = Student("Lenin", 13, 'M', 3)


# print(ravi)
# print(sneha)

# print(Student.__doc__)

ravi.setScores(10, 30)
total = ravi.getStudentTotalScore()
print("Ravi Total -", total)

sneha.setScores(30, 90)
snehaTotal = sneha.getStudentTotalScore()
print(snehaTotal)

rohit = Student("Rohit", 13, 'M', 4)

print(table)
```
